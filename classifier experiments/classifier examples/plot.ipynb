```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

```python
KN = pd.read_json('KNeighborsClassifier.json', orient='split')
DEC = pd.read_json('DecisionTreeClassifier.json', orient='split')
GRAD = pd.read_json('GradientBoostingClassifier.json', orient='split')
RAND = pd.read_json('RandomForestClassifier.json', orient='split')
SVC = pd.read_json('SVC.json', orient='split')
```

```python
KN.head()
```

```python
plt.scatter(KN.index, KN.Mean,label='KNeighborsClassifier')
plt.scatter(DEC.index, DEC.Mean,label='DecisionTreeClassifier')
plt.scatter(GRAD.index, GRAD.Mean,label='GradientBoostingClassifier')
plt.scatter(RAND.index, RAND.Mean,label='RandomForestClassifier')
plt.scatter(SVC.index, SVC.Mean,label='SVC')
plt.title('Indicators on N_600-D_4-Nmin_60-Zmin_0-CL_1_1-R_0')
plt.ylabel('Scores Mean')
plt.legend()
plt.grid()
plt.tight_layout()
plt.savefig('plot.png')
plt.show()
```
